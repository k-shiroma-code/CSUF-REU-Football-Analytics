```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report

#Load dataset
file_path = "combined_euros_1996_2024.csv"
df = pd.read_csv(file_path)

#Handling Missing Values
#df.fillna(method='ffill', inplace=True)  # Forward fill categorical values
#df.fillna(df.median(numeric_only=True), inplace=True)  # Fill numeric columns with median

#forward fill categorical values
df.ffill(inplace=True)

#fill numeric columns with median
df.fillna(df.select_dtypes(include=['number']).median(), inplace=True)

#ensure categorical columns are strings
df[['home_team', 'away_team', 'winner']] = df[['home_team', 'away_team', 'winner']].astype(str)

#convert 'winner' to match outcome instead of just team name
def determine_outcome(row):
    if row['winner'] == row['home_team']:
        return "home"
    elif row['winner'] == row['away_team']:
        return "away"
    else:
        return "draw"

df['match_outcome'] = df.apply(determine_outcome, axis=1)

#encode categorical variables
label_encoder = LabelEncoder()
df['home_team'] = label_encoder.fit_transform(df['home_team'])
df['away_team'] = label_encoder.fit_transform(df['away_team'])
df['match_outcome'] = label_encoder.fit_transform(df['match_outcome'])  # Encoding 'home', 'away', 'draw'

#feature engineer gd
df['goal_difference'] = df['home_score'] - df['away_score']

#check class distribution
print("\nClass Distribution in Target:")
print(df['match_outcome'].value_counts())

#define features and target
X = df[['home_team', 'away_team', 'goal_difference']]
y = df['match_outcome']

#train test split stratificiation 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

#train decision tree
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)
y_pred_dt = dt_model.predict(X_test)

print("\nDecision Tree Classifier Accuracy:", accuracy_score(y_test, y_pred_dt))
print("\nDecision Tree Classification Report:\n", classification_report(y_test, y_pred_dt, zero_division=1))

#train random forest 
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_test)

print("\nRandom Forest Classifier Accuracy:", accuracy_score(y_test, y_pred_rf))
print("\nRandom Forest Classification Report:\n", classification_report(y_test, y_pred_rf, zero_division=1))
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'max_depth': [3, 5, 10, None],
    'min_samples_split': [2, 5, 10]
}

grid_search = GridSearchCV(DecisionTreeClassifier(random_state=42), param_grid, cv=5)
grid_search.fit(X_train, y_train)
print("Best Parameters:", grid_search.best_params_)
```

```python
best_params = grid_search.best_params_

best_dt_model = DecisionTreeClassifier(**best_params, random_state=42)
best_dt_model.fit(X_train, y_train)

y_pred_best_dt = best_dt_model.predict(X_test)

print("\nOptimized Decision Tree Accuracy:", accuracy_score(y_test, y_pred_best_dt))
print("\nOptimized Decision Tree Classification Report:\n", classification_report(y_test, y_pred_best_dt, zero_division=1))
```

```python
results = pd.DataFrame(grid_search.cv_results_)

results["param_max_depth"] = results["param_max_depth"].astype(str)
results["param_min_samples_split"] = results["param_min_samples_split"].astype(str)

heatmap_data = results.pivot(index="param_max_depth", columns="param_min_samples_split", values="mean_test_score")

plt.figure(figsize=(8, 6))
sns.heatmap(heatmap_data, annot=True, fmt=".3f", cmap="coolwarm", linewidths=0.5)
plt.title("Hyperparameter Tuning - Decision Tree")
plt.xlabel("min_samples_split")
plt.ylabel("max_depth")
plt.show()
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Input
#from tensorflow.keras.models import Sequential
#from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical

#target variable to cataegorical 
y_train_cat = to_categorical(y_train)
y_test_cat = to_categorical(y_test)

input_dim = X_train.shape[1] 
output_dim = 3 

model = Sequential([
    Input(shape=(input_dim,)),  
    Dense(64, activation='relu'),
    Dropout(0.3),  
    Dense(32, activation='relu'),
    Dropout(0.2),
    Dense(output_dim, activation='softmax')  
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

history = model.fit(X_train, y_train_cat, epochs=50, batch_size=16, validation_data=(X_test, y_test_cat), verbose=1)

loss, accuracy = model.evaluate(X_test, y_test_cat)
print("\nNeural Network Accuracy:", accuracy)

y_pred_nn = model.predict(X_test)
y_pred_classes = y_pred_nn.argmax(axis=1)

print("\nNeural Network Classification Report:\n", classification_report(y_test, y_pred_classes, zero_division=1))
```
